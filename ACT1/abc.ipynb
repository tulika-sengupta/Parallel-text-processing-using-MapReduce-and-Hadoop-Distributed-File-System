```r
print(df$text)
```

```r
# build a corpus, and specify the source to be character vectors
myCorpus <- Corpus(VectorSource(df$text))
```

```r
dataframe <-data.frame(text=unlist(sapply(myCorpus, `[`)), stringsAsFactors=F)
print(dataframe$text)                                                    
```

```r
# remove URLs
removeURL <- function(x) gsub("?(f|ht)(tp)(s?)(://)(.*)[.|/](.*)", "", dataframe$text)
myCorpus <- tm_map(myCorpus, removeURL) 

# remove punctuation
myCorpus <- tm_map(myCorpus, removePunctuation) 
    
# remove numbers
myCorpus <- tm_map(myCorpus, removeNumbers)

myCorpusCopy <- myCorpus
# stem words
#myCorpus <- tm_map(myCorpus, stemDocument)
```

```r
dataframe <-data.frame(text=unlist(sapply(myCorpus, `[`)), stringsAsFactors=F)
print(dataframe$text)
```

```r
#removes non alphanumeric
myCorpus <- str_replace_all(dataframe$text, "[^[:alnum:]]", " ")

#remove smileys
myCorpus <- sapply(myCorpus,function(row) iconv(row, "latin1", "ASCII", sub=""))
```

```r
print(myCorpus)
```

```r
myCorpus <- Corpus(VectorSource(myCorpus))
myCorpus <- tm_map(myCorpus, content_transformer(tolower))
```

```r
dataframe <-data.frame(text=unlist(sapply(myCorpus, `[`)), stringsAsFactors=F)
print(dataframe$text)
```
